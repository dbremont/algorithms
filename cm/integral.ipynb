## Area Under a Curve

The **area under a curve** can be represented by the integral:


$A = \int_a^b f'(x) \, dx$


Where:
- $f(x)$ is the function representing the curve.
- $[a, b]$ is the interval over which the area is calculated.
- The function $f'(x) = x^2$ is used as an example.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function f(x)
def f_prima(x):
    return x**2

# Interval limits
a, b = 0, 2

# Compute the integral (area under the curve) using scipy
area, _ = quad(f_prima, a, b)

# Generate x values for plotting
x = np.linspace(a, b, 1000)
y = f(x)

# Plot the function and the area under the curve
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f\'(x) = x^2$', color='blue')
plt.fill_between(x, y, where=(x >= a) & (x <= b), color='lightblue', alpha=0.5, label=f'Area = {area:.2f}')

# Labels and title
plt.title('Area Under the Curve for $f\'(x) = x^2$')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Print the computed area
print(f"Area under the curve from x={a} to x={b}: {area:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function f(x)
def f_prima(x):
    return x**2

# Interval limits
a, b = 0, 2
n = 10  # Number of rectangles (Riemann sum)

# Compute the integral (area under the curve) using scipy
area, _ = quad(f_prima, a, b)

# Generate x values for plotting
x = np.linspace(a, b, 1000)
y = f_prima(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f_p(x) = x^2$', color='blue')

# Create rectangles for Riemann sum (left sum)
x_rect = np.linspace(a, b, n)
y_rect = f_prima(x_rect)
plt.bar(x_rect, y_rect, width=(b - a) / n, align='edge', color='orange', alpha=0.5, label='Riemann Sums (Rectangles)')

# Fill the area under the curve
plt.fill_between(x, y, where=(x >= a) & (x <= b), color='lightblue', alpha=0.5, label=f'Area = {area:.2f}')

# Labels and title
plt.title('Area Under the Curve for $f_p(x) = x^2$ with Riemann Sums')
plt.xlabel('x')
plt.ylabel('f_p(x)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Print the computed area
print(f"Area under the curve from x={a} to x={b}: {area:.2f}")
```
