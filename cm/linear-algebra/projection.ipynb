# Proyection

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define vectors v and u
v = np.array([3, 2])
u = np.array([2, 1])

# Compute projection of v onto u
def projection(v, u):
    return (np.dot(v, u) / np.dot(u, u)) * u

proj_v_on_u = projection(v, u)
```

```python
fig, ax = plt.subplots()
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v')
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='green', label='u')
ax.quiver(0, 0, proj_v_on_u[0], proj_v_on_u[1], angles='xy', scale_units='xy', scale=1, color='red', label='proj_v_on_u')

# Dashed line from projection tip to v tip
ax.plot([proj_v_on_u[0], v[0]], [proj_v_on_u[1], v[1]], 'k--', label='v - proj')

ax.set_xlim(-1, 4)
ax.set_ylim(-1, 3)
ax.set_aspect('equal')
ax.grid(True)
ax.legend()
ax.set_title("Projection of v onto u")

plt.show()
```
