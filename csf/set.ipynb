# Set

```python
import pylab as plt
from matplotlib_venn import venn3, venn3_circles
```

## Venn Diagram

```python
set1 = set(['A', 'B', 'C'])
set2 = set(['A', 'B', 'D'])
set3 = set(['A', 'E', 'F'])
```

```python
venn3([set1, set2, set3], ('Group1', 'Group2', 'Group3'))
plt.title("Venn Diagram: Letters Map to Numbers")
```
