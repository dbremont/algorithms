```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set parameters for the power law distribution
alpha = 2.5  # exponent parameter; typical values are between 1.5 and 3 for many real-world phenomena
xmin = 1     # minimum value of x

# Generate random values from a power law distribution
# Using inverse transform sampling
size = 10000  # number of data points
r = np.random.uniform(0, 1, size)
x = xmin * (1 - r) ** (-1 / (alpha - 1))

# Plotting the power law distribution
plt.figure(figsize=(8, 6))
plt.hist(x, bins=100, density=True, color='skyblue', edgecolor='black', log=True)
plt.xlabel('Value')
plt.ylabel('Probability Density (log scale)')
plt.title(f'Power Law Distribution (alpha = {alpha}, xmin = {xmin})')
plt.yscale('log')
plt.xscale('log')
plt.grid(True, which="both", linestyle='--', lw=0.5)
plt.show()
```
