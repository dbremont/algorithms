# Standard Deviation

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data: 100 data points from a normal distribution with mean=50, std=10
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # Sample standard deviation with n-1 in denominator

print(f"Mean: {mean:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")

# Plot histogram of the data
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7, density=True)

# Plot mean line
plt.axvline(mean, color='red', linestyle='-', linewidth=2, label='Mean')

# Plot ±1 std dev lines
plt.axvline(mean + std_dev, color='orange', linestyle='--', linewidth=2, label='+1 Std Dev')
plt.axvline(mean - std_dev, color='orange', linestyle='--', linewidth=2, label='-1 Std Dev')

# Plot ±2 std dev lines
plt.axvline(mean + 2*std_dev, color='green', linestyle=':', linewidth=2, label='+2 Std Dev')
plt.axvline(mean - 2*std_dev, color='green', linestyle=':', linewidth=2, label='-2 Std Dev')

plt.title('Data Distribution with Mean and Standard Deviations')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
