# Pearson Correlation Coefficient

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [2, 4, 6, 8]
Y = [1, 3, 2, 10]
```

```python
r_numpy = np.corrcoef(X, Y)[0, 1]
```

```python

plt.scatter(X, Y)
plt.title(f"Correlation Coefficient: {r_numpy:.2f}")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```
