# Covariance

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data with covariance
np.random.seed(0)
mean = [5, 5]
cov = [[3, 1], [1, 2]]  # Covariance matrix
data = np.random.multivariate_normal(mean, cov, 500)

# Plot the data
plt.scatter(data[:, 0], data[:, 1], alpha=0.5, label='Data Points')

# Covariance ellipse
eigvals, eigvecs = np.linalg.eig(cov)
angle = np.arctan2(eigvecs[1, 0], eigvecs[0, 0])
width, height = 2 * np.sqrt(eigvals)

# Create the ellipse
ellipse = plt.matplotlib.patches.Ellipse(mean, width, height, angle=np.degrees(angle), edgecolor='r', facecolor='none', lw=2, label='Covariance Ellipse')
plt.gca().add_patch(ellipse)

# Show the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Covariance Error Ellipse and Data')
plt.legend()
plt.grid(True)
plt.show()
```
