# Multi-armed Bandit Problem

```python
import numpy as np
```

```python

# Define the bandit problem
np.random.seed(42)
n_arms = 3  # Number of bandit arms
true_rewards = [0.2, 0.5, 0.8]  # True reward probabilities for each arm
n_steps = 1000  # Number of actions to take

# Initialize variables
estimated_rewards = np.zeros(n_arms)  # Initial estimates for rewards
action_counts = np.zeros(n_arms)  # Count of times each arm is pulled
epsilon = 0.1  # Exploration probability

# Store rewards for plotting
rewards = []

# Reinforcement Learning loop
for step in range(n_steps):
    # Choose an action using Îµ-greedy policy
    if np.random.rand() < epsilon:  # Explore
        action = np.random.randint(0, n_arms)
    else:  # Exploit
        action = np.argmax(estimated_rewards)
    
    # Simulate pulling the arm (reward is 1 with the arm's true probability)
    reward = 1 if np.random.rand() < true_rewards[action] else 0
    rewards.append(reward)
    
    # Update action counts and reward estimates
    action_counts[action] += 1
    estimated_rewards[action] += (reward - estimated_rewards[action]) / action_counts[action]
```

```python
# Results
print(f"True rewards: {true_rewards}")
print(f"Estimated rewards: {estimated_rewards}")
print(f"Action counts: {action_counts}")
```

```python
# Plotting results
import matplotlib.pyplot as plt

plt.plot(np.cumsum(rewards), label="Cumulative Reward")
plt.xlabel("Steps")
plt.ylabel("Cumulative Reward")
plt.title("Multi-Armed Bandit: Cumulative Reward")
plt.legend()
plt.show()
```
