# Gradient Descent

```python
import numpy as np
```

```python
# Generate a simple dataset
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # Feature (100 samples, 1 feature)
y = 4 + 3 * X + np.random.randn(100, 1)  # Target with noise

# Initialize parameters
m, b = 0.0, 0.0  # Initial guess for slope and intercept
learning_rate = 0.1
iterations = 1000
n = len(X)

# Gradient Descent Algorithm
for i in range(iterations):
    # Predictions
    y_pred = m * X + b
    
    # Calculate gradients
    gradient_m = -(2/n) * np.sum(X * (y - y_pred))  # Partial derivative w.r.t. m
    gradient_b = -(2/n) * np.sum(y - y_pred)        # Partial derivative w.r.t. b
    
    # Update parameters
    m -= learning_rate * gradient_m
    b -= learning_rate * gradient_b
    
    # Print loss every 100 iterations
    if i % 100 == 0:
        loss = np.mean((y - y_pred)**2)
        print(f"Iteration {i}: Loss = {loss:.4f}")
```

```python
print(f"Final parameters: m = {m:.4f}, b = {b:.4f}")

# Predictions using the trained model
import matplotlib.pyplot as plt

plt.scatter(X, y, color="blue", label="Data")
plt.plot(X, m * X + b, color="red", label="Best Fit Line")
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.title("Gradient Descent Linear Regression")
plt.show()
```
